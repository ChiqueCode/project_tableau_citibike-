```python
# Dependancies
import pandas as pd
import numpy as np
```

```python
# CSV file 
csv_file_june = "Resources/june.csv"
df_june = pd.read_csv(csv_file_june)
df_june.head()
```

```python
df_june.shape[0]
```

```python
csv_file_july = "Resources/july.csv"
df_july = pd.read_csv(csv_file_july)
df_july.head()
```

```python
df_july.shape[0]
```

```python
csv_file_august = "Resources/august.csv"
df_august = pd.read_csv(csv_file_august)
df_august.head()
```

```python
df_august.shape[0]
```

```python
vertical_stack = pd.concat([df_june, df_july, df_august], axis=0)
```

```python
vertical_stack.head(100)
```

```python
vertical_stack.shape[0]
```

```python
# Drop the NaN values so it won't affect the analysis
vertical_stack_df = vertical_stack.dropna(axis=1, how='all')
vertical_stack_df.head(100)
```

```python
vertical_stack_df.shape[0]
```

```python
vertical_stack_df.isnull().sum(axis = 0)
```

```python
vertical_stack_df.to_csv(index=False)
```
